```python
# Import dependencies
import pandas as pd
# pd.options.display.float_format = '{:.0f}'.format
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
# from datetime import datetime 
```

```python
# Import csv files
departments = pd.read_csv("../Resources/departments.csv")
dept_emp = pd.read_csv("../Resources/dept_emp.csv")
dept_manager = pd.read_csv("../Resources/dept_manager.csv")
employees = pd.read_csv("../Resources/employees.csv")
salaries = pd.read_csv("../Resources/salaries.csv")
titles = pd.read_csv("../Resources/titles.csv")
```

```python
# Convert data into pandas dataframes
departments_df = pd.DataFrame(departments)
dept_emp_df = pd.DataFrame(dept_emp)
dept_manager_df = pd.DataFrame(dept_manager)
employees_df = pd.DataFrame(employees)
salaries_df = pd.DataFrame(salaries)
titles_df = pd.DataFrame(titles)
```

## Find the count of employees by department ordered alphabetically by department name 

```python
# View departments dataframe
departments_df.head()
```

```python
# View dept_emp dataframe
dept_emp_df.head()
```

```python
# Merge departments and dept_emp dataframes
dept_emp_dept_merged_df = pd.merge(departments_df, dept_emp_df, on=["dept_no", "dept_no"])
dept_emp_dept_merged_df.head()
```

```python
# Group by dept_name and count employees in each department
dept_emp_dept_grouped = dept_emp_dept_merged_df.groupby(["dept_name"]).count()["emp_no"].sort_values(ascending=False)
dept_emp_dept_grouped_df = pd.DataFrame(dept_emp_dept_grouped)
dept_emp_dept_grouped_df
```

```python
# Generate a bar plot showing the total number of employees by department 



dept_emp_dept_grouped_df.plot(kind="bar", ylabel="Total Employees", legend=None)
plt.title("Number of Employees by Department")
plt.xlabel("Department")
plt.show()




```

```python
# Convert all data in Years from 'object' to 'float' data type

employees_df['hire_date'] = pd.to_datetime(employees_df['hire_date'])
        
```

```python
# Check to see that the data type for hire_date was successfully changed to datetime

employees_df.dtypes
```

```python
# Group infomation by Year of hire_date and count the employees hired each year
employees_df_grouped = employees_df.groupby((employees_df["hire_date"]).dt.year).count()["emp_no"]
employees_df_grouped_df = pd.DataFrame(employees_df_grouped)
employees_df_grouped_df.reset_index()

```

```python
# Create a plot of the number of employees hired each year

plt.figure(figsize=(12,5))

x = range(1985, 2001)
plt.plot(x, employees_df_grouped_df["emp_no"], marker="o")

plt.title("Number of New Hires per Year:  1985 - 2000")
plt.xlabel('Year')
plt.ylabel('New Hires')

plt.show()


# plt.savefig("xx/xx.png")
```

## Find the Average and Median salary by Title

```python
# View the salaries dataframe
salaries_df
```

```python
# View the titles dataframe
titles_df
```

```python
# Merge the employees and the salaries dataframes
emp_salaries_merge = pd.merge(employees_df, salaries_df, on= ["emp_no", "emp_no"])
emp_salaries_merge.head()
```

```python
# Rename column "emp_title_id" to "title_id"


emp_salaries_merge_rename = emp_salaries_merge.rename(columns={'emp_no' : 'emp_no',
                                                                'emp_title_id' : 'title_id',
                                                                'birth_date' : 'birth_date',
                                                                'first_name' : 'first_name',
                                                                'last_name' : 'last_name',
                                                                'sex' : 'sex',
                                                                'hire_date' : 'hire_date',
                                                                'salary' : 'salary'                                                               
})


emp_salaries_merge_rename.head()
```

```python
# Merge emp_salaries_merge and the titles dataframes
emp_sal_title_merge = pd.merge(emp_salaries_merge_rename, titles_df, on= ["title_id", "title_id"])
emp_sal_title_merge
```

```python
# Group by title and find the average salary for each title
emp_sal_title_merge_grouped = emp_sal_title_merge.groupby(["title"]).mean()["salary"]
emp_sal_title_merge_grouped_df = pd.DataFrame(emp_sal_title_merge_grouped)
emp_sal_title_merge_grouped_df = emp_sal_title_merge_grouped_df.rename(columns={'title' : 'title',
                                                                'salary' : 'mean_salary'})

emp_sal_title_merge_grouped_df.reset_index()

```

```python
# Plot bar graph of saverage alary by title

emp_sal_title_merge_grouped_df.plot(kind="bar", ylabel="Average Salary", legend=None)
plt.title("Average Salary by Job Title")
plt.xlabel("Title")

plt.show()
```

```python
# Find the median salary by job title

# Group by title and find the average salary for each title
emp_sal_title_merge_grouped_med = emp_sal_title_merge.groupby(["title"]).median()["salary"]
emp_sal_title_merge_grouped_med_df = pd.DataFrame(emp_sal_title_merge_grouped_med)
emp_sal_title_merge_grouped_med_df = emp_sal_title_merge_grouped_med_df.rename(columns={'title' : 'title',
                                                                'salary' : 'median_salary'})
emp_sal_title_merge_grouped_med_df.reset_index()
```

```python
# Plot bar graph of median salary by title

emp_sal_title_merge_grouped_med_df.plot(kind="bar", ylabel="Median Salary", legend=None)
plt.title("Median Salary by Job Title")
plt.xlabel("Title")

plt.show()
```

### Find the total salaries by department and plot in a bar chart and a pie chart

```python
# Combine existing merged emp_salaries_merge_rename dataframe with dept_emp dataframe
emp_sal_dept_emp_merge = pd.merge(emp_salaries_merge_rename, dept_emp, on= ["emp_no", "emp_no"])
emp_sal_dept_emp_merge_df = pd.DataFrame(emp_sal_dept_emp_merge)
emp_sal_dept_emp_merge_df
```

```python
# Combine existing merged emp_sal_dept_emp_merge dataframe with departments dataframe
emp_sal_dept_emp_dept_merge = pd.merge(emp_sal_dept_emp_merge, departments_df, on= ["dept_no", "dept_no"])
emp_sal_dept_emp_dept_merge_df = pd.DataFrame(emp_sal_dept_emp_dept_merge)
emp_sal_dept_emp_dept_merge_df
```

```python
# Group by department and total the salary for each department
emp_sal_dept_emp_dept_group = emp_sal_dept_emp_dept_merge_df.groupby(['dept_name']).sum()['salary']
emp_sal_dept_emp_dept_group_df = pd.DataFrame(emp_sal_dept_emp_dept_group)
emp_sal_dept_emp_dept_group_df.reset_index()
```

```python
# Plot bar graph of median salary by title

emp_sal_dept_emp_dept_group_df.plot(kind="bar", ylabel="Total Salary", legend=None)
plt.title("Total Salary by Department")
plt.xlabel("Department")

plt.show()
```

```python
# Plot pie chart of total salary by department
emp_sal_dept_emp_dept_group_df.plot(kind='pie', subplots=True, figsize=(10, 8), autopct="%1.1f%%")
plt.title("Total Salary by Department")
# plt.legend(loc=(1.1,1))
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.ylabel("")
plt.show()
```
